# K-Means Clustering

```python
from sklearn.cluster import KMeans
```

```python
fig=plt.figure(figsize=(14,8))
wcss=[]
for i in range(1,11):
    clf=KMeans(n_clusters=i)
    clf.fit(crime_df)
    wcss.append(clf.inertia_)
plt.plot(range(1,11),wcss)
plt.title('Elbow method')
plt.ylabel('WCSS')
plt.xlabel('no of clusters')
plt.show()
```

```python
clf=KMeans(n_clusters=4)
clf
```

```python
k_cluster = KMeans(random_state=42)
k_cluster.fit(crime_df)
```

```python
y=pd.DataFrame(k_cluster.fit_predict(crime_df),columns=['clusterid_kmeans'])
y['clusterid_kmeans'].value_counts() 
```

```python
crime['clusterid_kmeans']=k_cluster.labels_
crime 
```

```python
crime.groupby('clusterid_kmeans').agg(['mean']).reset_index()
```

```python
fig,ax=plt.subplots(figsize=(14,8))
crime.groupby(['clusterid_kmeans']).count()['location'].plot(kind='bar')
plt.ylabel(['location'])
plt.title('KMeans clustering')
ax.set_xlabel('Cluster')
ax.set_ylabel('location')
#plt.yticks(fontsize=15)
#plt.xtricks(fontsize=15)
plt.show()
```
